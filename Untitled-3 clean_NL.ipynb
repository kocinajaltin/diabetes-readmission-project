```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

# MySQL credentials
username = "root"
password = "1234"
host = "localhost"
port = "3306"
database = "diabetes_schema"

# Create SQLAlchemy engine
engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}")

# Query: Readmission rate by average age
query_age_avg = """
SELECT 
    p.age_avg,
    COUNT(*) AS total,
    SUM(CASE WHEN readmitted = '<30' THEN 1 ELSE 0 END) AS readmitted_30
FROM 
    encounters e
LEFT JOIN
    patients p
ON e.patient_nbr = p.patient_nbr
GROUP BY 
    p.age_avg
ORDER BY 
    p.age_avg;
"""

# Query: Readmission rate by gender
query_gender = """
SELECT 
    p.gender,
    COUNT(*) AS total,
    SUM(CASE WHEN e.readmitted = '<30' THEN 1 ELSE 0 END) AS readmitted_30
FROM 
    encounters e
LEFT JOIN 
    patients p ON e.patient_nbr = p.patient_nbr
GROUP BY 
    p.gender;
"""

# Load into DataFrames
df_age = pd.read_sql(query_age_avg, engine)
df_gender = pd.read_sql(query_gender, engine)

# Compute readmission rates
df_age["readmission_rate"] = df_age["readmitted_30"] / df_age["total"]
df_gender["readmission_rate"] = df_gender["readmitted_30"] / df_gender["total"]

# Plot
plt.figure(figsize=(12, 5))

# Age_avg plot
plt.subplot(1, 2, 1)
plt.bar(df_age["age_avg"], df_age["readmission_rate"], color='steelblue')
plt.xticks(rotation=45)
plt.title("Readmission Rate by Average Age")
plt.ylabel("Rate")
plt.xlabel("Average Age")

# Gender plot
plt.subplot(1, 2, 2)
plt.bar(df_gender["gender"], df_gender["readmission_rate"], color='orange')
plt.title("Readmission Rate by Gender")
plt.ylabel("Rate")
plt.xlabel("Gender")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import xgboost as xgb
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
import matplotlib.pyplot as plt

df = pd.read_csv('C:/Users/ak20adb/OneDrive - University of Hertfordshire/Pictures/portfolio/test1.csv')

print("Original dataset info:")
print(f"Shape: {df.shape}")
print(f"Columns: {list(df.columns)}")

print("\nOriginal readmitted value distribution:")
print(df["readmitted"].value_counts())

# Create binary target: 1 for <30 days readmission, 0 for everything else
# Assuming your 0,1,2 values map to: 0=NO, 1=<30, 2=>30
df["readmitted_30day"] = (df["readmitted"] == 1).astype(int)

print(f"\nBinary target distribution:")
print(f"Not readmitted within 30 days (0): {sum(df['readmitted_30day'] == 0)}")
print(f"Readmitted within 30 days (1): {sum(df['readmitted_30day'] == 1)}")

# Check for missing values BEFORE dropping
print(f"\nMissing values per column:")
missing_info = df.isnull().sum()
print(missing_info[missing_info > 0])

print(f"\nTotal missing values: {df.isnull().sum().sum()}")
print(f"Rows with ANY missing values: {df.isnull().any(axis=1).sum()}")
print(f"Rows with ALL values present: {df.dropna().shape[0]}")

# Instead of dropping ALL rows with ANY missing values, let's be more selective
print(f"\nBefore cleaning: {df.shape[0]} rows")

# Option 1: Drop rows only if target is missing
df_clean = df.dropna(subset=['readmitted_30day'])
print(f"After dropping rows with missing target: {df_clean.shape[0]} rows")

# Option 2: Handle missing values more intelligently
# Fill missing values for specific columns instead of dropping entire rows
df_filled = df.copy()

# Example: Fill missing numeric columns with median
numeric_cols = df_filled.select_dtypes(include=['int64', 'float64']).columns
for col in numeric_cols:
    if col not in ['readmitted', 'readmitted_30day']:
        df_filled[col] = df_filled[col].fillna(df_filled[col].median())

# Fill missing categorical columns with mode or 'Unknown'
categorical_cols = df_filled.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df_filled[col] = df_filled[col].fillna('Unknown')

print(f"After intelligent filling: {df_filled.shape[0]} rows")

# Use the filled dataset
df = df_filled

# Feature/target split
X = df.drop(columns=["readmitted", "readmitted_30day"])
X = pd.get_dummies(X, drop_first=True)
y = df["readmitted_30day"]

print(f"\nFinal dataset for modeling:")
print(f"Features shape: {X.shape}")
print(f"Target shape: {y.shape}")
print(f"Target distribution: {y.value_counts()}")

# Only proceed if we have data
if len(X) > 0 and len(y) > 0:
    # Train/test split with stratification
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, stratify=y, test_size=0.2, random_state=42
    )
    
    print(f"\nDataset sizes:")
    print(f"Training: {X_train.shape[0]} samples")
    print(f"Testing: {X_test.shape[0]} samples")
    print(f"Features: {X_train.shape[1]}")
    
    # Continue with model training...
    model = xgb.XGBClassifier(
        objective='binary:logistic',
        scale_pos_weight=sum(y_train == 0) / sum(y_train == 1),
        random_state=42,
        eval_metric='auc'
    )
    
    model.fit(X_train, y_train)
    
    # Make predictions and evaluate
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1]
    
    print(f"\nAUC-ROC Score: {roc_auc_score(y_test, y_pred_proba):.3f}")
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
    
else:
    print("ERROR: No data remaining after cleaning!")
    print("Check your data file and missing value handling strategy.")
    
```
